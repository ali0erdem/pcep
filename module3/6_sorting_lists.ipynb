### The bubble sort ###





```python
# The bubble sort
my_list = [8, 10, 6, 2, 4]  # list to sort
swapped = True  # It's a little fake, we need it to enter the while loop.

while swapped:
    swapped = False  # no swaps so far
    for i in range(len(my_list) - 1):
        if my_list[i] > my_list[i + 1]:
            swapped = True  # a swap occurred!
            my_list[i], my_list[i + 1] = my_list[i + 1], my_list[i]

print(my_list)
```

```python
# This can explain what's going on under the hood. Much easier to undersand!
my_list = [8, 10, 6, 2, 4]  # list to sort
swapped = True  # It's a little fake, we need it to enter the while loop.
count = 0
while swapped:
    swapped = False  # no swaps so far
    for i in range(len(my_list) - 1):
        print(i, f"{my_list[i]} > {my_list[i + 1]}")
        if my_list[i] > my_list[i + 1]:
            count += 1
            swapped = True  # a swap occurred!
            my_list[i], my_list[i + 1] = my_list[i + 1], my_list[i]
            print(f"swap number:{count} - list: {my_list}")
    print(f"after for loop swap: {swapped}")

print(my_list)
```

```python
# The bubble sort - interactive version
my_list = []
swapped = True
num = int(input("How many elements do you want to sort: "))

for i in range(num):
    val = float(input("Enter a list element: "))
    my_list.append(val)

while swapped:
    swapped = False
    for i in range(len(my_list) - 1):
        if my_list[i] > my_list[i + 1]:
            swapped = True
            my_list[i], my_list[i + 1] = my_list[i + 1], my_list[i]

print("\nSorted:")
print(my_list)
```
