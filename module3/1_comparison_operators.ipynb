### Comparison: equality operator ###

- To ask the question, -are two values equal?- you use the `== (equal equal)` operator.
- `a = b` **assigns** a with the value of b; `a == b` **compares** a and b.
- It is a `binary operator(needs two arguments) with left-sided binding`.

```python
print(2 == 2)
print(2 == 2.0)
print(1 == 2)
```

```python
var = 1  # Assigning 1 to var
print(var == 0)
```

### Comparison: Inequality operator ###

- The `!= (not equal to)` operator compares the values of two operands.
- If they are `not equal`, the result of the comparison is `True`.

```python
var = 1  # Assigning 1 to var
print(var != 0)
```
